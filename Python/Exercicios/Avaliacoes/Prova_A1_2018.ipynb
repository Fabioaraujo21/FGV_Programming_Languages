Seja d(n) definido como a suma dos divisores inteiros de n. Se d(a) = b e d(b) = a, sendo a ≠ b, então a e b são um "par amigável" de números.  

Por exemplo, os divisores inteiros de 220 são 1, 2, 4, 5, 10, 11, 20, 22, 44, 55 e 110; então d(220) = 284.  
Os divisores inteiros de 284 são 1, 2, 4, 71 e 142; então d(284) = 220.  

Encontre todos os "pares amigáveis" de números menores que 20000.  Crie uma lista de tuplas com esses pares.  

```python
def divisores(n):
    lista=[]
    for i in range(1,n):
        if n%i==0:
            lista.append(i)
    return lista
def somadiv(n):
    return sum(divisores(n))
paresamigaveis1=[]
paresamigaveis2=[]

for i in range(1,20000):     ###Demora muito com valores grandes
    for j in range(1,20000):
        if somadiv(i)==j and somadiv(j)==i:
            paresamigaveis1.append(i)
            paresamigaveis2.append(j) 
print(paresamigaveis1)
print(paresamigaveis2)
#lista de tuplas
tupla=tuple(zip(paresamigaveis1,paresamigaveis2))
tupla
```
